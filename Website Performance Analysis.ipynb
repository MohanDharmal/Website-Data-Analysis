```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WebData.csv')
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
```

```python

df = df.drop(index=0).reset_index(drop=True)
```

```python
df.head()
```

```python
df.columns = ["Channel Group","DateHour","Users","Sessions","Engaged sessions","Average engagement time per session","Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df.head()
```

```python
df.info()
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"],format='%Y%m%d%H',errors='coerce') 
```

```python
df.info()
```

```python
df.head()
```

```python
num_cols = df.columns.drop(["Channel Group","DateHour"])
df[num_cols]= df[num_cols].apply(pd.to_numeric,errors='coerce')
df.info()
```

```python
df['Hour'] = df['DateHour'].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

# Session and User over Time

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10,5))
plt.title("Session and Users Over Time")
plt.xlabel("DateHour")
plt.ylabel("Count")
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
```

# Total Users By Channels

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Channel Group',y='Users',estimator=np.sum)
plt.title('Total Usres By Channels')
plt.xticks(rotation=45)
plt.show()
```

# Average Engagement Time By Channel

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Channel Group',y='Average engagement time per session',estimator=np.mean)
plt.xticks(rotation=45)
plt.show()
```

# Engagement Rate Distribution By Channels

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='Channel Group',y='Engagement rate')
plt.xticks(rotation=45)
plt.show()
```

# Engaged VS Non Engaged Sessions

```python
session_df = df.groupby('Channel Group')[['Sessions','Engaged sessions']].sum().reset_index()
session_df['Non-Engaged'] = session_df['Sessions'] - session_df['Engaged sessions']
session_df_melted = session_df.melt(id_vars='Channel Group',value_vars=['Engaged sessions','Non-Engaged'])
session_df_melted
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted,x='Channel Group',y='value',hue='variable')
plt.title("Engaged VS Non Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

# Traffic by Hour and Channel

```python
heatmap_data = df.groupby(["Hour","Channel Group"])['Sessions'].sum().unstack().fillna(0)
heatmap_data
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(heatmap_data,linewidth=.5,annot=True,fmt='.0f')
plt.title('Traffic By Hour and Channel')
plt.show()
```
